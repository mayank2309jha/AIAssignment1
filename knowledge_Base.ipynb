```python
# Now we give weather to these locations
import random
extreme_temperature = ["Extreme","Mild"]
weather = ["Summer", "Winter"]
cloudy = ["Cloudy","Not Cloudy"]
rain = ["Rain","No Rain"]

# Generate a random forecast
arr = []
for i in range(0,len(df)):
    random_forecast = [
        random.choice(extreme_temperature),
        random.choice(weather),
        random.choice(cloudy),
        random.choice(rain)
    ]
    # Convert the forecast to a string
    weather_forecast = ', '.join(random_forecast)
    arr.append(weather_forecast)
df["Weather"] = arr
```
